```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta
import polars as pl
import implicit
import lightgbm as lgb
from itertools import groupby
import faiss
from tqdm import tqdm

from utils import create_features, recall_at, fit_lgb_ranker
```

```python
df_test_users = pl.read_parquet(f'test_users.pq') # пересчитать тотал с ними
df_clickstream = pl.read_parquet(f'clickstream.pq')

df_cat_features = pl.read_parquet(f'cat_features.pq')
df_text_features = pl.read_parquet(f'text_features_norm.pq')
df_event = pl.read_parquet(f'events.pq')
```

```python
treshold = df_clickstream['event_date'].max() #- timedelta(days=7)
```

```python
df_train_17 = (
    df_clickstream
    .filter(df_clickstream['event_date']<= treshold)
    .filter(pl.col('event')==17)
    .join(df_event, on='event', how='left')
    .join(df_text_features, on='item', how='inner')
    .sort(['cookie', 'event_date'])
    .group_by('cookie')
    .tail(5)
    .with_columns(pl.col("event_date").dt.offset_by("-1y").alias("event_date"))
)
```

```python
df_train_non17 = (
    df_clickstream
    .filter(df_clickstream['event_date']<= treshold)
    .filter(pl.col('event')!=17)
    .join(df_event, on='event', how='left')
    .join(df_text_features, on='item', how='inner')
)
```

```python
df_train = pl.concat([df_train_17, df_train_non17], how="vertical").sort(['cookie', 'event_date'])
```

```python
import polars as pl
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
import random

active_users = (
    df_train.lazy()
    .group_by("cookie")
    .agg(pl.count().alias("item_count"))
    .filter(pl.col("item_count") >= 100)
    .filter(pl.col("item_count") <= 200)
    .collect()
    .get_column("cookie")
    .to_list()
)

for plot_num in range(10):
    plt.figure(figsize=(10, 8))
    
    selected_users = random.sample(active_users, 10) # [14419, 15735, 37892, 71842, 142621, 112616, 109585, 46014]
    
    colors = plt.cm.tab10(np.linspace(0, 1, 10))
    
    all_vectors = np.array(
        df_train.filter(pl.col("cookie").is_in(selected_users))
        .get_column("title_projection")
        .to_list(),
        dtype=np.int8
    )
    pca = PCA(n_components=2).fit(all_vectors)
    
    for user_idx, user in enumerate(selected_users):
        user_df = df_train.filter(pl.col("cookie") == user)
        vectors = np.array(user_df.get_column("title_projection").to_list(), dtype=np.int8)
        points = pca.transform(vectors)
        
        user_str = str(user)
        display_name = f"{user_str[:6]}..." if len(user_str) > 6 else user_str
        
        plt.scatter(
            points[:, 0], 
            points[:, 1], 
            color=colors[user_idx], 
            label=f"User {display_name} ({len(points)} items)",
            alpha=0.6,
            s=40
        )
    
    plt.title(f"Item Title Vectors (PCA) - Random User Comparison {plot_num+1}/10")
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(alpha=0.2)
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import polars as pl
from sklearn.cluster import MiniBatchKMeans
import faiss

df_items = (
    df_train
    .select(['item', 'title_projection', 'node']).unique() # add items with contact last week
)
items = df_items['item'].to_list()
title_vectors = np.array(df_items['title_projection'].to_list(), dtype=np.float32)

nodes_for_items = df_items['node'].to_list()
item_to_node = dict(zip(items, nodes_for_items))

index = faiss.IndexFlatIP(64)
index.add(title_vectors)
```

```python
user_groups = df_train.group_by('cookie').agg([pl.col('title_projection'), pl.col('node')])

user_centroids = {}
user_last_item = {}
user_seen_nodes = {}

for cookie, title_projs, nodes in tqdm(user_groups.iter_rows(), total=len(user_groups), desc="Processing user groups"):
    user_vecs = np.array(title_projs, dtype=np.float32)
    centroid = np.mean(user_vecs, axis=0)
    user_centroids[cookie] = [centroid]
    user_last_item[cookie] = [user_vecs[-1]]
    user_seen_nodes[cookie] = set(nodes)
```

```python
from collections import defaultdict
import heapq

recommendations = {}
user_cookies = list(user_centroids.keys())
batch_size = 2000

for i in tqdm(range(0, len(user_cookies), batch_size), desc="Batch processing"):
    batch_cookies = user_cookies[i:i+batch_size]
    all_centroids = []
    user_offsets = []
    current_offset = 0
    
    for cookie in batch_cookies:
        centroids = user_centroids[cookie]
        all_centroids.extend(centroids)
        user_offsets.append((current_offset, current_offset + len(centroids)))
        current_offset += len(centroids)
    
    centroid_matrix = np.vstack(all_centroids)
    scores, indices = index.search(centroid_matrix, 2000)
    for j, cookie in enumerate(batch_cookies):
        start, end = user_offsets[j]
        user_scores = scores[start:end]
        user_indices = indices[start:end]
        
        top_items = defaultdict(float)
        for centroid_scores, centroid_items in zip(user_scores, user_indices):
            for score, idx in zip(centroid_scores, centroid_items):
                item = items[idx]
                if score > top_items.get(item, -1):
                    top_items[item] = score
        
        top_200 = heapq.nlargest(2000, top_items.items(), key=lambda x: x[1])
        seen_nodes = user_seen_nodes.get(cookie, set())
        recommended_nodes = []
        added_nodes = set()
        
        for item, _ in top_200:
            node = item_to_node.get(item)
            if node and node not in seen_nodes and node not in added_nodes:
                recommended_nodes.append(node)
                added_nodes.add(node)
                if len(recommended_nodes) >= 200:
                    break
        
        recommendations[cookie] = recommended_nodes
```

```python
def predict_like(cookie: str, item_vector: np.ndarray) -> float:
    """Returns the maximum similarity score between item and user's centroids."""
    centroids = user_centroids.get(cookie, [])
    if not centroids:
        return 0.0  # Handle new users appropriately
    
    # faiss.normalize_L2(item_vector.reshape(1, -1))
    similarities = [np.dot(centroid, item_vector) for centroid in centroids]
    return max(similarities)
```

```python
df_recommendations = (
    pl.DataFrame({
        "cookie": list(recommendations.keys()),
        "recommended_items": list(recommendations.values())
    })
    .explode("recommended_items")
    .with_columns(
        pl.col("recommended_items").alias("node"),
        pl.int_range(0, pl.len()).over("cookie").alias("rank") + 1
    )
    .filter(pl.col("rank") <= 200)
    .select(["cookie", "node", "rank"])
)
```

```python
df_recommendations.write_csv('retrieval_data/cluster_text_pred.csv')
```
